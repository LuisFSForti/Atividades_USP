```python
import matplotlib.pyplot as plt
import numpy as np
import math
import tabulate as tb
```

```python
#Para calcular lambda e quantas meias-vidas sero processadas
def CalcularDecaimento(valorInicial, meiaVida, tempos):
    k = valorInicial
    #lambda = -(ln(valorInicial/2) - ln(k))/meiaVida = ln(2)/meiaVida
    lam = math.log(2)/meiaVida

    meiasVidas = []
    for t in tempos:
        meiasVidas.append(t/meiaVida)

    return k, lam, meiasVidas

#Quantos pontos vai calcular
qtdVal = int(1e04)
#Comea com t = 0 e da cria uma distribuio logartmica, para ter menos pontos na assntota final das curvas
tempos = [0]
tempos.extend(np.logspace(0, math.log(48 * 3600, 10), qtdVal-1))

#Meia vida sem segundos
meiaVida1, meiaVida2 = 3600, 3600*5

#Calcula os lambdas e as meias-vidas
k1, lam1, meiasVidas1 = CalcularDecaimento(100, meiaVida1, tempos)
k2, lam2, meiasVidas2 = CalcularDecaimento(0, meiaVida2, tempos)
#O terceiro no tem meia-vida, pois  estvel
k3, lam3 = 0, 0

#Calcula as constantes  fim das contas terem o valor correto em t=0
c1 = k1
c2 = -k1 * lam1/(lam1 - lam2) - k2
c3 = k1 + k2 + k3

#Armazenam a quantidade de ncelos dos elementos
qtds1 = []
qtds2 = []
qtds3 = []

#Para cada t
for t in tempos:
    #Calcula x, y e z no instante t
    x = c1 * math.pow(math.e, -lam1 * t)
    y = -((c1 * lam1 * math.pow(math.e, -lam1 * t))/(lam1 - lam2)) - c2 * math.pow(math.e, -lam2 * t)
    z = c1 * lam2 * math.pow(math.e, -lam1 * t)/(lam1 - lam2) + c2 * math.pow(math.e, -lam2 * t) + c3

    qtds1.append(x)
    qtds2.append(y)
    qtds3.append(z)

#Transforma o tempo de segundos para horas
tempos = [x / 3600 for x in tempos]

#Imprime os dados em forma numrica
print(tb.tabulate(np.column_stack(([k1,k2,k3], [lam1, lam2, lam3], [meiaVida1, meiaVida2, 0], [c1,c2,c3])), 
                  headers=["Elemento","Valor inicial","Lambda","Meia-vida (s)","Ci"], tablefmt="github", showindex="always"))
print("\n\n\n")

#Imprime os grficos
fig, ax = plt.subplots(3,2)

fig.set_figheight(15)
fig.set_figwidth(15)

ax[0,0].plot(tempos, qtds1)
ax[0,0].grid()
ax[0,0].set_xlabel('Tempo (H)')
ax[0,0].set_ylabel('Quantidade (N)')
ax[0,0].title.set_text('Elemento 1 em funo de t')

ax[0,1].plot(meiasVidas1, qtds1)
ax[0,1].grid()
ax[0,1].set_xlabel('Meias-vidas')
ax[0,1].set_ylabel('Quantidade (N)')
ax[0,1].title.set_text('Elemento 1 em funo de meias-vidas')

ax[1,0].plot(tempos, qtds2)
ax[1,0].grid()
ax[1,0].set_xlabel('Tempo (H)')
ax[1,0].set_ylabel('Quantidade (N)')
ax[1,0].title.set_text('Elemento 2 em funo de t')

ax[1,1].plot(meiasVidas2, qtds2)
ax[1,1].grid()
ax[1,1].set_xlabel('Meias-vidas')
ax[1,1].set_ylabel('Quantidade (N)')
ax[1,1].title.set_text('Elemento 2 em funo de meias-vidas')

ax[2,0].plot(tempos, qtds3)
ax[2,0].grid()
ax[2,0].set_xlabel('Tempo (H)')
ax[2,0].set_ylabel('Quantidade (N)')
ax[2,0].title.set_text('Elemento 3 em funo de t')

ax[2,1].plot(tempos, qtds1, label="X")
ax[2,1].plot(tempos, qtds2, label="Y")
ax[2,1].plot(tempos, qtds3, label="Z")
ax[2,1].grid()
ax[2,1].set_xlabel('Tempo (H)')
ax[2,1].set_ylabel('Quantidade (N)')
ax[2,1].title.set_text('Todos os elementos em funo de t')
ax[2,1].legend()

fig.tight_layout()
plt.show()
```
