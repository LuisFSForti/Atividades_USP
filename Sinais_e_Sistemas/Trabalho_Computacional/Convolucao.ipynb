Neste trabalho foi analizado como calcular a convolução entre duas funções quaisquer funcA e funcB e como animar o processo do cálculo. O projeto foi realizado em Jupyter Notebook por meio da linguagem Python. Os gráficos foram feitos com a biblioteca matplotlib.

Inicialmente, foi criada a função para processar a convolução. Ela requer duas funções e seus nomes, além de um vetor de valores de tempo onde deveria calcular. Assim ela pode criar os dois gráficos superiores, os quais mostram as funções dadas no intervalo.

Então, para calcular a convolução, foi feita uma análise da fórmula.

```python
from IPython.display import Image
from IPython.core.display import HTML 
Image(url= 
      "https://wikimedia.org/api/rest_v1/media/math/render/svg/7206fe9d1b6e0ad341016c78c3f939e0d3c1d14e")
```

Criou-se um loop para passar por cada instante de tempo do intervalo escolhido, onde esse seria o instante τ da convolução. Recalcula-se os valores de funcB, deslocando-a em relação ao t. Como τ está negativo para a funcB e deslocada, ela deve então estar espelhada em relação ao ponto t. Ao multiplicar os valores das duas funções, obtém-se os valores que seráo utilizados para o cálculo da convolução total. Estes valores se tornaram o gráfico inferior direito das imagens, demonstrando os valores da convolução no instante τ.

A integral de uma função é definida como a área definida pela mesma no intervalo. Por esta definição, pode-se interpretar os valores calculados como a altura de retângulos e, se calculada a largura entre cada valor, pode-se encontrar a área dos mesmos.

```python
Image(url= 
      "https://upload.wikimedia.org/wikipedia/commons/thumb/4/4a/Integral_approximations_J.svg/220px-Integral_approximations_J.svg.png")
```

Se o vetor de tempo for uniformemente distribuído, então também serão os valores calculados. Assim, pode-se definir que a distância entre dois valores é dada pelo comprimento total de tempo dividido pela quantidade de valores. Assim pode-se calcular a integral da convolução no instante t.

$$
  \int_{a}^b funcA(τ) \cdot funcB(t -τ) dτ \approx \sum_{τ = a}^b funcA(τ) \cdot funcB(t - τ) \cdot \frac{tempo[tempo.size - 1] - tempo[0]}{tempo.size}
$$

Assim pode-se calcular a convolução total entre duas funções quaisquer, a qual foi desenhada como o gráfico inferior esquerdo.

Neste trabalho foram analisadas algumas convoluções e sua propriedade de comutação.

# 1. Código

```python
import numpy as np
import matplotlib.pylab as plt
#Funções necessárias para atualizar gráficos no Jupyter Notebook
from IPython.display import display, clear_output

#Plota o gráfico das funções e sua convolução
def PlotConvolucao(funcA, nomeFuncA, funcB, nomeFuncB, tempo):
    #Cria o plot
    fig, ax = plt.subplots(2,2)
    fig.set_figheight(10)
    fig.set_figwidth(10)

    #Cria as variáveis para armazenar os valores das funções
    valA = np.zeros(shape=(tempo.size))
    valB = np.zeros(shape=(tempo.size))
    valConv = np.zeros(shape=(tempo.size))

    #Para cada t em tempo
    for i, t in enumerate(tempo):
            #Calcula e salva o valor das funções
            valA[i] = funcA(t)
            valB[i] = funcB(t)

    #Faz o plot das funções
    ax[0,0].plot(tempo, valA, "-b", label="f1 = " + nomeFuncA)
    ax[0,0].grid()
    
    ax[0,1].plot(tempo, valB, "-r", label="f2 = " + nomeFuncB)
    ax[0,1].grid()

    #Para cada instante tau
    for posTau, tau in enumerate(tempo):
        #Calcula a função b com o deslocamento de tau
        for i in range(0, tempo.size):
            #Tempo negativo para espelhar a função
            valB[i] = funcB(-(tempo[i] - tau))

        #Calcula os valores da convolução entre as funções no instante
        conv = valA * valB
        
        #integral = área no intervalo definido = somatório das áreas
        #Como conv tem os valores no instante, que equivalem à altura, basta multiplicar pelo comprimento de cada trecho
        #(tempo[tempo.size - 1] - tempo[0]) = comprimento total
        #tempo.size = quantidade de trechos
        #(tempo[tempo.size - 1] - tempo[0])/tempo.size = comprimento de cada trecho
        #np.sum(conv) * (tempo[tempo.size - 1] - tempo[0])/tempo.size = área total no instante
        valConv[posTau] = np.sum(conv) * (tempo[tempo.size - 1] - tempo[0])/tempo.size

        #Limpa o plot da convolução
        ax[1,0].cla()
        #Plota a função A
        ax[1,0].plot(tempo, valA, "-b")
        #Plota a função B no instante atual
        ax[1,0].plot(tempo, valB, "-r")
        #Plota os valores da convolução calculados até o momento
        ax[1,0].plot(tempo, valConv, "-g", label="f1 ∗ f2")
        #Linha tracejada vertical que marca o instante atual
        ax[1,0].axvline(x = tau, color = 'r', ls='--', label = 'τ')
        ax[1,0].grid()

        #Limpa o plot da convolução no instante atual
        ax[1,1].cla()
        #Plota os valores da convolução no momento atual
        ax[1,1].plot(tempo, conv, "-g", label="Convolução em τ")
        ax[1,1].grid()

        #Ativa as legendas, todas no canto superior direito
        ax[0,0].legend(loc=1)
        ax[0,1].legend(loc=1)
        ax[1,0].legend(loc=1)
        ax[1,1].legend(loc=1)

        #Desenha os gráficos
        display(fig)

        #Fala para limpar o gráfico quando houver um novo para desenhar
        clear_output(wait = True)
        #Espera um pequeno intervalo de tempo, para a animação ocorrer em, no mínimio, 5 segundos
        plt.pause(5/tempo.size)
```

# 2. Funções utilizadas

```python
#Rect(t)
def Rect(t):
    if abs(t) > 0.5:
        return 0
    return 1

#sin(t)
def Seno(t):
    return np.sin(t)

#t/3 * Rect(t/3 - 1/2)
def MeioTriang(t):
    return t/3 * Rect(t/3 - 1/2)

#t/3 * sin(t) * Rect(t/3 - 1/2)
def MeioTriangSenoidal(t):
    return Seno(t) * MeioTriang(t)
```

# 3.1 Exemplo 1 - Duas funções retangulares

Para o primeiro teste, foram utilizadas duas funções retangulares de 1 de largura, as quais geraram um triângulo de altura 1 e largura 2, assim como esperado.

Aqui notou-se algo: a operação é extremamente lenta. Após pesquisar, descobriu-se que a biblioteca matplotlib não é eficiente para apresentação em tempo real (Fonte: https://stackoverflow.com/questions/8955869/why-is-plotting-with-matplotlib-so-slow), no entanto, por falta de tempo e como a falha não afeta os resultados, não foi corrigida.

```python
t = 2
tempo = np.linspace(-t, t, 500)

PlotConvolucao(Rect, "Rect(t)", Rect, "Rect(t)", tempo)
```

# 3.2 Exemplo 2 - Função retangular e seno

Para o segundo teste, foram utilizadas uma função retangular de 1 de largura e uma função seno. O resultado foi uma função seno com amplitude ligeriamente menor, assim como o esperado.

```python
t = 2*np.pi
tempo = np.linspace(-t, t, 500)

PlotConvolucao(Rect, "Rect(t)", Seno, "Sin(t)", tempo)
```

# 3.4 Exemplo 4 - Função customizada e seno

Para os testes finais, foi criada uma outra função customizada pela multiplicação da função anterior com a seno e foi então calculada sua convolução com outra função seno.
