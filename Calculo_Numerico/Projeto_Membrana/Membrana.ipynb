```python
import numpy as np
import scipy
import matplotlib.pylab as plt
from mpl_toolkits import mplot3d
from matplotlib import cm

import time
import tabulate as tb
import matplotlib.pyplot as pltT
```

## 1.Código do professor

```python
# Função para plotar os curvas de nível constante de temperatura
def PlotaMembrane(N1,N2,L1,L2,W):
    x = np.linspace(0, L1, N1)
    y = np.linspace(0, L2, N2)
    X,Y=np.meshgrid(x,y)
    Z = np.copy(W)
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set(xlabel='x', ylabel='y', title='Deslocamento vertical')
    im = ax.contourf(X, Y, Z, 20)
    im2 = ax.contour(X, Y, Z, 20, linewidths=0.25, colors='k')
    fig.colorbar(im, ax=ax)
    plt.show()
    
def PlotaSurface(N1,N2,L1,L2,W):
    x = np.linspace(0, L1, N1)
    y = np.linspace(0, L2, N2)
    X,Y=np.meshgrid(x,y)
    Z = np.copy(W)
    ax = plt.axes(projection ='3d')
    ax.plot_surface(X, Y, Z,cmap='viridis', edgecolor='none')
    ax.set(xlabel='x', ylabel='y', title='Deslocamento vertical')
    ax.set_zlim(-1.01, 1.01)
    ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False)

def ij2n (i, j, N):
    return i + j*N

def BuildMatrizesEigen(N1, N2, sigma, rho, e, delta):
    nunk = N1*N2

    # Stiffness matrix K: Build it as a sparse matrix 
    d1 = 4.0*np.ones(nunk)
    d2 = -np.ones(nunk-1)
    d3 = -np.ones(nunk-N1)
    K = (sigma/delta**2)*scipy.sparse.diags([d3, d2, d1, d2, d3], [-N1, -1, 0, 1, N1], format='csr')

    # Force the eigenvalues associated to boundary points 
    # to be a big number as compared to fundamental modes
    big_number = 10000
    Iden = big_number*scipy.sparse.identity(nunk, format='csr')

    # Lados verticais
    for k in range(0,N2):
        Ic = ij2n(0,k,N1) # Left
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]

        Ic = ij2n(N1-1,k,N1) # Right
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]
        
    # Lados horizontais
    for k in range(0,N1):
        Ic = ij2n(k,0,N1) # Bottom
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]

        Ic = ij2n(k,N2-1,N1) # Top
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]

    # Mass matrix: Simple case, multiple of identity
    M = rho*e*scipy.sparse.identity(nunk, format='csr')
    
    return K, M    
```

## 2. Código do projeto

```python
#Para construir uma membrana de lados iguais e mesma quantidade de pontos em cada eixo
#Possui pouquíssima diferença do código do professor, apenas mudando o fato de usar um único N
def BuildMembranaQuadrada(N, sigma, rho, e, delta):
    nunk = N**2

    # Stiffness matrix K: Build it as a sparse matrix 
    d1 = 4.0*np.ones(nunk)
    d2 = -np.ones(nunk-1)
    d3 = -np.ones(nunk-N)
    K = (sigma/delta**2)*scipy.sparse.diags([d3, d2, d1, d2, d3], [-N, -1, 0, 1, N], format='csr')

    big_number = 10000
    Iden = big_number*scipy.sparse.identity(nunk, format='csr')

    # Lados verticais
    for k in range(0,N):
        Ic = ij2n(0,k,N) # Left
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]

        Ic = ij2n(N-1,k,N) # Right
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]
        
    # Lados horizontais
    for k in range(0,N):
        Ic = ij2n(k,0,N) # Bottom
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]

        Ic = ij2n(k,N-1,N) # Top
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]

    # Mass matrix: Simple case, multiple of identity
    M = rho*e*scipy.sparse.identity(nunk, format='csr')
    
    return K, M

#Dados da membrana quadrada
rhoQ = 1
sigmaQ = 1.0
eQ = 1
LQ = 1.0

#Tamanhos de grade utilizados
tamanhosQ = [11, 12, 13, 14, 15, 16, 18, 21, 23, 26, 31, 36, 41, 51, 61, 81, 101, 121, 151]
#Frequências obtidas
frequenciasQ = []
#Tempos obtidos
temposQ = []

#Para cada tamanho de grade
for N in tamanhosQ:
    #Distancia entre pontos, vale para os eixos X e Y
    delta = LQ / (N-1) 
        
    #Aplicar restrições nas bordas
    KQ, MQ = BuildMatrizesEigen(N, N, sigmaQ, rhoQ, eQ, delta)

    #Tempo no começo do cálculo
    tempo = time.time()
    
    #Método tipo potencias: Apenas os modos fundamentais (trabalha com esparsas!)
    #Calcula para os primeiros 4 pontos
    LamQ, QQ = scipy.sparse.linalg.eigsh(KQ, k=4, M=MQ, which='SM')

    #Pega as frequências nos pontos
    omegasQ = np.sqrt(LamQ)

    #Salva o tempo que levou para calcular e as frequências obtidas
    temposQ.append(time.time()-tempo)
    frequenciasQ.append(omegasQ)

#Salva as frequências em forma de array do numpy
frequenciasQ = np.array(frequenciasQ)
```

```python
#Para construir uma membrana de lados iguais, mas cortada na diagonal, assim formando um tiângulo
def BuildMembranaTriangular(N, sigma, rho, e, delta):
    nunk = N**2

    # Stiffness matrix K: Build it as a sparse matrix 
    d1 = 4.0*np.ones(nunk)
    d2 = -np.ones(nunk-1)
    d3 = -np.ones(nunk-N)
    K = (sigma/delta**2)*scipy.sparse.diags([d3, d2, d1, d2, d3], [-N, -1, 0, 1, N], format='csr')

    big_number = 10000
    Iden = big_number*scipy.sparse.identity(nunk, format='csr')

    # Lados verticais
    for k in range(0,N):
        Ic = ij2n(0,k,N) # Left
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]

        Ic = ij2n(N-1,k,N) # Right
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]
        
    # Lados horizontais
    for k in range(0,N):
        Ic = ij2n(k,0,N) # Bottom
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]

        Ic = ij2n(k,N-1,N) # Top
        K[Ic,:], K[:,Ic] = Iden[Ic,:], Iden[:,Ic]

    # Mass matrix: Simple case, multiple of identity
    M = rho*e*scipy.sparse.identity(nunk, format='csr')

    #Para cortar o gráfico no meio
    #Para cada ponto no eixo X
    for i in range(N):
        #Se for o primeiro ponto em X (canto inferior esquerdo), então ele deve ser o único da coluna
        if i == 0:
            #Para cada ponto no eixo Y
            for j in range(1, N):
                #Calcula a posição dele
                Ic = ij2n(i,j,N)
                #Define que ele deve estar estático
                K[Ic,:] = Iden[Ic,:]

            #Avança para o próximo X
            continue

        #Se não for o primeiro ponto no eixo X
        #Para cada ponto no eixo Y
        for j in range(N):
            #Se a tangente for maior que 1 (ângulo maior que 45°, portanto fora do triângulo)
            if j/i > 1:
                #Calcula a posição dele
                Ic = ij2n(i,j,N)
                #Define que ele deve estar estático
                K[Ic,:] = Iden[Ic,:]
    
    return K, M

#Dados da membrana triangular
LT = 1
sigmaT = 1.0
rhoT = 1
eT = 1

#Tamanhos de grade utilizados
tamanhosT = [11, 12, 13, 14, 15, 16, 18, 21, 23, 26, 31, 36, 41, 51, 61, 81, 101, 121, 151]
#Frequências obtidas
frequenciasT = []
#Tempos obtidos
temposT = []

#Para cada tamanho de grade
for N in tamanhosT:
    #Distancia entre pontos, vale para os eixos X e Y
    delta = LT / (N-1) 
        
    # Aplicar restrições nas bordas
    KT, MT = BuildMembranaTriangular(N, sigmaT, rhoT, eT, delta)

    #Tempo no começo do cálculo
    tempo = time.time()
    
    # Método tipo potencias: Apenas os modos fundamentais (trabalha com esparsas!)
    #Calcula para os primeiros 4 pontos
    LamT, QT = scipy.sparse.linalg.eigsh(KT, k=4, M=MT, which='SM')

    #Pega as frequências nos pontos
    omegasT = np.sqrt(LamT)

    #Salva o tempo que levou para calcular e as frequências obtidas
    temposT.append(time.time()-tempo)
    frequenciasT.append(omegasT)


#Salva as frequências em forma de array do numpy
frequenciasT = np.array(frequenciasT)
```
