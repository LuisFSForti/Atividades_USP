Classe de Grafos:

```python
import numpy as np

class Aresta:
    def __init__(self, nos, info):
        self._Nos = nos
        self._Info = info

    def SetInfo(self, info):
        self._Info = info

    def SetNosConectados(self, nos):
        self._Nos = nos
    
    def GetInfo(self):
        return self._Info

    def GetNosConectados(self):
        return self._Nos

class No:
    def __init__(self, nome, info):
        self._Nome = nome
        self._Info = info
        self._Conexoes = []

    def ConectarNo(self, no, infoAresta):
        self._Conexoes.append(Aresta([self, no], infoAresta))
        no._Conexoes.append(Aresta([no, self], infoAresta))

    def DesconectarLigacao(self, no, nos):
        for aresta in self._Conexoes:
            if aresta.GetNosConectados()[1] == no:
                self._Conexoes.remove(aresta)
                break

        for aresta in no._Conexoes:
            if aresta.GetNosConectados()[1] == self:
                no._Conexoes.remove(aresta)
                break

        if len(no._Conexoes) < 1:
            nos.remove(no)
            print(no.GetNome(), " ficou sem conexoes, foi removido")
            del no
        
        if len(self._Conexoes) < 1:
            nos.remove(self)
            print(self.GetNome(), " ficou sem conexoes, foi removido")
            del self

    def EstaConectadoCom(self, no):
        for conexao in self._Conexoes:
            if conexao.GetNosConectados()[1] == no:
                return True

        return False

    def SetInfo(self, info):
        self._Info = info
    
    def GetInfo(self):
        return self._Info

    def GetNome(self):
        return self._Nome

    def GetConexoes(self):
        return self._Conexoes

class Grafo:
    def __init__(self):
        self._Nos = []

    def CriarNo(self, nome, info):
        self._Nos.append(No(nome, info))

    def EncontrarPorNome(self, nome):
        for no in self._Nos:
            if no.GetNome() == nome:
                return no

        return None
    
    def ConectarNosPorNome(self, nome1, nome2, infoAresta):
        no1 = self.EncontrarPorNome(nome1)
        no2 = self.EncontrarPorNome(nome2)
        if not no1:
            print("Nome ", nome1, " inexistente!")
            return

        if not no2:
            print("Nome ", nome2, " inexistente!")
            return

        if not no1.EstaConectadoCom(no2):
            no1.ConectarNo(no2, infoAresta)
        else:
            print("Ja estavam conectados os nos", no1.GetNome(), no2.GetNome(), sep=" ")

    def DesconectarNosPorNome(self, nome1, nome2):
        no1 = self.EncontrarPorNome(nome1)
        no2 = self.EncontrarPorNome(nome2)

        if not no1:
            print("Nome ", nome1, " inexistente!")
            return

        if not no2:
            print("Nome ", nome2, " inexistente!")
            return

        no1.DesconectarLigacao(no2, self._Nos)

    def ExcluiNo(self, nome):
        no = self.EncontrarPorNome(nome)

        if not no:
            print("Nome ", nome, " inexistente!")
            return
        
        for conexao in list(reversed(no.GetConexoes())):
            no.DesconectarLigacao(conexao.GetNosConectados()[1], self._Nos)

    def Imprimir(self):
        for no in self._Nos:
            for aresta in no.GetConexoes():
                print([conectado.GetNome() for conectado in aresta.GetNosConectados()], aresta.GetInfo())

    def MontarMatriz(self):
        matriz = np.zeros((len(self._Nos),len(self._Nos)))
        
        for no in self._Nos:
            for conexao in no.GetConexoes():
                matriz[self._Nos.index(no)][self._Nos.index(conexao.GetNosConectados()[1])] = 1

        return matriz
                
            
grafo = Grafo()

grafo.CriarNo("Tubo1", "")
grafo.CriarNo(2, "")
grafo.ConectarNosPorNome("Tubo1",2, 100)
grafo.ConectarNosPorNome("Tubo1",2, -100)

grafo.Imprimir()
```
