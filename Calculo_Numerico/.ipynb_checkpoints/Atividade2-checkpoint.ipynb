```python
#Exercício 1
import numpy as np

def Assembly(conec, C, nv):
    nc = len(C)
    A = np.zeros(shape=(nv,nv))
    for k in range(nc):
        n1 = conec[k, 0] - 1
        n2 = conec[k, 1] - 1
        
        A[n1, n1] += C[k]
        A[n1, n2] += -C[k]
        A[n2, n1] += -C[k]
        A[n2, n2] += C[k]

    return A

nv = 5
conec = np.zeros(shape=(7,2), dtype=int)

conec[0] = [1,2]
conec[1] = [2,3]
conec[2] = [3,4]
conec[3] = [4,5]
conec[4] = [5,2]
conec[5] = [5,3]
conec[6] = [5,1]

C = [2,2,1,2,1,2,2]

print("A = ", Assembly(conec, C, nv))
```

```python
#Exercício 2
#Importante: é necessário rodar a célula anterior previamente!

def SolveNetwork(conec, C, nv, natm, nB, QB, imprimir):
    Atilde = Assembly(conec, C, nv)
    Atilde[natm-1, :] = 0
    Atilde[natm-1, natm-1] = 1
    
    b = np.zeros(shape=nv)
    b[nB-1] = QB

    pressure = np.linalg.solve(Atilde, b)

    if imprimir:
        print("Ã = ", Atilde, "\n\nb = ", b, "\n\nPressão = ", pressure)
    return pressure

pressao = SolveNetwork(conec, C, nv, 3, 1, 3, True)
```

```python
#Exercício 3
#Importante: é necessário rodar as células anteriores previamente!

def FormarK(C, nv):
    nc = len(C)
    K = np.zeros(shape=(nc,nc))
    for k in range(nc):
        K[k, k] = C[k]

    return K

def FormarD(conec, C, nv):
    nc = len(C)
    D = np.zeros(shape=(nc,nv))
    for k in range(nc):
        for j in range(nv):
            n1 = conec[k, 0] - 1
            n2 = conec[k, 1] - 1

            if j == n1:
                D[k, j] = 1
            else: 
                if j == n2:
                    D[k, j] = -1

    return D

p = SolveNetwork(conec, C, nv, 3, 1, 3, True)
K = FormarK(C, nv)
D = FormarD(conec, C, nv)
Q = K.dot(D).dot(p)

print("\nK = ", K)
print("\nD = ", D)
print("\nQ = ", Q)

Vazões = abs(Q)
Variacao = [-Q[0] + Q[6], Q[0] -Q[1] + Q[4], Q[1] - Q[2] + Q[5], Q[2] - Q[3], Q[3] - Q[4] - Q[5] - Q[6]]
print("Variação por cano = ", Variacao)
print("Variação total = ", sum(Variacao))
```

```python
#Exercício 4
#Importante: é necessário rodar as células anteriores previamente!

def CalcularPotencia(p, D, K):
    pT = np.transpose(p)
    Dt = np.transpose(D)
    aux = Dt.dot(K).dot(D)
    W = pT.dot(aux).dot(p)
    return W

W = CalcularPotencia(p, D, K)
print("Potencia = ", W)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def GeraRede(nx,ny,CH,CV):
    nv = nx*ny
    nc = (nx-1)*ny + (ny-1)*nx
    
    coord = np.zeros(shape=(nv,2))

    for i in range(nx):
      for j in range(ny):
        ig = i + j*nx
        coord[ig,0] = i
        coord[ig,1] = j

    conec = np.zeros(shape=(nc,2), dtype=int)
    C = np.zeros(nc)
    
    # Loop sobre canos horizontais
    for j in range(ny):
        for i in range(nx-1):
          k = j*(nx-1) + i
          conec[k,0] = j*nx + i
          conec[k,1] = j*nx + i+1
          C[k] = CH

    # Loop sobre canos verticais
    for i in range(nx):
      for j in range(ny-1):
          k = (nx-1)*ny + j*nx + i
          conec[k,0] = i + j*nx
          conec[k,1] = i + (j+1)*nx
          C[k] = CV

    return nv, nc, conec, C, coord
```

```python
def PlotPressure(nx, ny, pressure):
    x = np.arange(0, nx, 1)
    y = np.arange(0, ny, 1)
    X,Y=np.meshgrid(x,y)
    Z = np.copy(pressure)
    Z.shape = (ny,nx)
    fig, ax = plt.subplots(figsize=(10,10))
    ax.set_aspect('equal')
    ax.set(xlabel='x', ylabel='y', title='Contours of pressure')
    ax.grid()
    ax.set_xticks(x)
    ax.set_yticks(y)
    cp = plt.contourf(X, Y, Z, cmap='jet')
    cb = plt.colorbar(cp)
    plt.show()

def PlotPressureAsSurface(nx, ny, pressure):
    x = np.arange(0, nx, 1)
    y = np.arange(0, ny, 1)
    X,Y=np.meshgrid(x,y)
    Z = np.copy(pressure)
    Z.shape = (ny,nx)
    fig, ax = plt.subplots(figsize=(10,10))
    ax.set_aspect('equal')
    ax.set(xlabel='x', ylabel='y', title='Contours of pressure')
    ax.grid()
    ax.set_xticks(x)
    ax.set_yticks(y)
    ax = plt.axes(projection="3d")
    ax.plot_surface(X, Y, Z, cmap='jet', edgecolor=None)
    ax.set(xlabel='$x$', ylabel='$y$', zlabel='$z$')
    ax.view_init(elev=30., azim=45)
    plt.show()
```

```python
#Exercício 5
#Importante: é necessário rodar as células anteriores previamente!

def CalcularRedeQuadrada(CH, CV, nx, ny, nAtm, nB, QB, desenhar):
    nv, nc, conec, C, coord = GeraRede(nx, ny , CH, CV)

    #Este trecho é para correção dos valores, pois o código dos outros exercícios foi adaptado para o primeiro nó ser o 1, não o 0
    for i in range(len(conec)):
        conec[i][0] += 1
        conec[i][1] += 1
                
    p = SolveNetwork(conec, C, nv, nAtm, nB, QB, False)
    
    if desenhar:
        PlotPressure(nx, ny, p)
        PlotPressureAsSurface(nx, ny, p)

CH = 1.3
CV = 3.6
nx, ny = 12, 10
CalcularRedeQuadrada(CH, CV, nx, ny, nx*ny, 1, 3, True)
CalcularRedeQuadrada(CH, CV, nx, ny, nx*ny - (int)(nx/2) - (int)(ny/2*nx), 1, 3, True)
```

```python
#Exercício 6
#Importante: é necessário rodar as células anteriores previamente!

import time

tempos =[]
tamanhos = np.logspace(start=1.31, stop=2.256, num=50, dtype=int)
CH = 1.3
CV = 3.6

for i in tamanhos:
    tempo = time.time()
    CalcularRedeQuadrada(CH, CV, i, i, i*i, 1, 3, False)
    tempos.append(time.time() - tempo)

fig, ax = plt.subplots(1)
ax.plot(tamanhos, tempos, marker='o')
ax.loglog()
ax.legend(['Vetor loglog'])
ax.grid()
ax.set_ylabel('t (s)')
ax.set_xlabel('tamanho')

plt.show()

fig, ax = plt.subplots(1)
ax.plot(tamanhos, tempos)
ax.loglog()
ax.legend(['Vetor loglog'])
ax.grid()
ax.set_ylabel('t (s)')
ax.set_xlabel('tamanho')

plt.show()
```
