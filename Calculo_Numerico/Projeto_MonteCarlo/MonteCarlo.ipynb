```python
def GeraRede(nx,ny,CH,CV):
    nv = nx*ny
    nc = (nx-1)*ny + (ny-1)*nx
    
    coord = np.zeros(shape=(nv,2))

    for i in range(nx):
      for j in range(ny):
        ig = i + j*nx
        coord[ig,0] = i
        coord[ig,1] = j

    conec = np.zeros(shape=(nc,2), dtype=int)
    C = np.zeros(nc)
    
    # Loop sobre canos horizontais
    for j in range(ny):
        for i in range(nx-1):
          k = j*(nx-1) + i
          conec[k,0] = j*nx + i
          conec[k,1] = j*nx + i+1
          C[k] = CH

    # Loop sobre canos verticais
    for i in range(nx):
      for j in range(ny-1):
          k = (nx-1)*ny + j*nx + i
          conec[k,0] = i + j*nx
          conec[k,1] = i + (j+1)*nx
          C[k] = CV

    return nv, nc, conec, C, coord

def PlotPressure(nx, ny, pressure):
    x = np.arange(0, nx, 1)
    y = np.arange(0, ny, 1)
    X,Y=np.meshgrid(x,y)
    Z = np.copy(pressure)
    Z.shape = (ny,nx)
    fig, ax = plt.subplots(figsize=(10,10))
    ax.set_aspect('equal')
    ax.set(xlabel='x', ylabel='y', title='Contours of pressure')
    ax.grid()
    ax.set_xticks(x)
    ax.set_yticks(y)
    cp = plt.contourf(X, Y, Z, cmap='jet')
    cb = plt.colorbar(cp)
    plt.show()

def PlotPressureAsSurface(nx, ny, pressure):
    x = np.arange(0, nx, 1)
    y = np.arange(0, ny, 1)
    X,Y=np.meshgrid(x,y)
    Z = np.copy(pressure)
    Z.shape = (ny,nx)
    fig, ax = plt.subplots(figsize=(10,10))
    ax.set_aspect('equal')
    ax.set(xlabel='x', ylabel='y', title='Contours of pressure')
    ax.grid()
    ax.set_xticks(x)
    ax.set_yticks(y)
    ax = plt.axes(projection="3d")
    ax.plot_surface(X, Y, Z, cmap='jet', edgecolor=None)
    ax.set(xlabel='$x$', ylabel='$y$', zlabel='$z$')
    ax.view_init(elev=30., azim=45)
    plt.show()
```

## 2. Código da lista

```python
#Código necessário

import numpy as np
import matplotlib.pyplot as plt

#Montar a tabela A
def Assembly(conec, C):
    #Procura o maior valor entre o maior valor de cada coluna, assim encontrando a quantidade de nós
    nv = max([max(conec[:,0]), max(conec[:,1])])
    #Quantidade de canos = número de ligações = tamanho da lista de Condutâncias
    nc = len(C)
    
    A = np.zeros(shape=(nv,nv))
    #Para cada cano
    for k in range(nc):
        #Pega os nós conectados
        n1 = conec[k, 0] - 1
        n2 = conec[k, 1] - 1

        #Salva na matriz A os valores da condutância do cano de acordo com os nós respectivos
        A[n1, n1] += C[k]
        A[n1, n2] += -C[k]
        A[n2, n1] += -C[k]
        A[n2, n2] += C[k]

    return A

#Para calcular as pressões de cada nó
def SolveNetwork(conec, C, natm, nB, QB, imprimir):
    #Procura o maior valor entre o maior valor de cada coluna, assim encontrando a quantidade de nós
    nv = max([max(conec[:,0]), max(conec[:,1])])

    #Pega a matriz A
    Atilde = Assembly(conec, C)
    #Altera a linha do nó conectado à atmosfera
    Atilde[natm-1, :] = 0
    Atilde[natm-1, natm-1] = 1

    #Cria a matriz de variação da massa
    b = np.zeros(shape=nv)
    #Salva em qual nó está entrando massa externa
    b[nB-1] = QB

    #Resolve a matriz, assumindo Ã como dados, b como resultados e calculando o vetor de pressões intermediário
    pressure = np.linalg.solve(Atilde, b)

    #Para não desacelerar quando não é necessário
    if imprimir:
        print("Ã = ", Atilde, "\n\nb = ", b, "\n\nPressão = ", pressure)
        
    return pressure
```

## 3. Código do projeto

```python
#Entupir canos aleatórios
def RandomFailFinos(C, po, Centup):
    #Copia C para não alterá-lo indevidamente
    Cnew = np.copy(C)
    #Pega a quantidade da canos
    nc = len(C)
    #Para cada cano
    for k in range(nc):
        #Valor entre 0 e 1
        x = np.random.rand()
        #Se valor sorteado for menor ou igual à probabilidade de entupir
        #&&
        #For um cano fino
        if x <= po and Cnew[k] == 2:
            #Salva o cano como entupido
            Cnew[k] = Centup

    return Cnew

#Para calcular a probabilidade de ter uma pressão elevada em função da quantidade de realizações, probabilidade de entupir e a nova condutância
def CalcularProbabilidade(nx, ny, conec, C, qtdCasos, po, Centup):
    qtdCasosPressao = 0

    #Para cada caso
    for i in range(int(qtdCasos)):
        #Aleatoriza os canos entupidos
        Cnovo = RandomFailFinos(C, po, Centup)
        #Calcula as pressões
        p = SolveNetwork(conec, Cnovo, nx*ny, 1, 10, False)
        #Se algum nó ficou com pressão maior ou igual a 12
        if max(p) >= 12:
            #Salva o ocorrido
            qtdCasosPressao += 1

    #Retorna a porcentagem de casos em que houve pressão maior ou igual a 12 dentre o total de casos
    return qtdCasosPressao/qtdCasos
```

```python
###Formar a rede base

#Dados iniciais
CH = 2.0
CV = 2.0
nx, ny = 10, 10
nv, nc, conec, C, coord = GeraRede(nx, ny , CH, CV)

#Canos grossos
lst = [0, 1, 2, 30, 31, 32, 93, 103, 113]
C[lst] = 20.0

#O código foi projetado para o primeiro nó ser o 1, então aumenta o valor de todos em 1
conec[:][:] += 1
```

```python
###Para calcular por po

#Para pegar os valores de po
pOs = np.linspace(start=0, stop=1, num=100, dtype=float)
probabilidadesPorPo = []

#Calcula as probabilidades (quantidade de realizações = 5*10^3)
for i in pOs:
    probabilidadesPorPo.append(CalcularProbabilidade(nx, ny, conec, C, 5*10**3, i, 0.2))

#Para deixá-los em forma de %
pOs *= 100
probabilidadesPorPo = [i * 100 for i in probabilidadesPorPo]
```

```python
###Para calcular pela quantidade de realizações

#Para pegar as quantidades de realizações
#Foi dada mais ênfase aos primeiros valores, pois é onde a variação é maior
qtd = np.concatenate((np.linspace(start=1, stop=500, num=500, dtype=int),np.linspace(start=501, stop=10**4, num=50, dtype=int)))
probabilidadesPorQtd = []

#Calcula as probabilidades (po = 0.1)
for i in qtd:
    probabilidadesPorQtd.append(CalcularProbabilidade(nx, ny, conec, C, i, 0.1, 0.2))

#Para deixá-los em forma de %
probabilidadesPorQtd = [i * 100 for i in probabilidadesPorQtd]
```

## 4. Resultados

```python
###Exemplo de rede entupida

#Aleatoriza os canos entupidos (50% de chance)
Cnovo = RandomFailFinos(C, 0.1, 0.2)
#Calcula as pressões
p = SolveNetwork(conec, Cnovo, nx*ny, 1, 10, False)

PlotPressure(nx, ny, p)
PlotPressureAsSurface(nx, ny, p)
```
